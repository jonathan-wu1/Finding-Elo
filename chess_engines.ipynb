```python
import pandas as pd
from IPython.display import set_matplotlib_formats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/landing/computer-chess-ability.csv")
```

```python
df.head()
```

```python
years = list(df["Year"])
elo_ratings = list(df["Elo rating"])
```

```python
# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(8, 5))

# Scatter plot and line plot

ax.plot(years, elo_ratings, color='#332288', linestyle='-', marker = 'v', 
        alpha = 0.5, label='Highest Rated Chess Engine')

# Add a horizontal line at y = 2800
ax.axhline(y=2882, color='#CC6677', linestyle='--', label='Current Highest Rated Human Player')

# Set labels and title
ax.set_title('Chess Engine Ratings Over Time', pad=10,  color='#333333', weight='bold')
ax.set_xlabel('Year', labelpad=10, color='#333333', fontsize = 12)
ax.set_ylabel('Elo Rating', labelpad=10, color='#333333', fontsize = 12)

ax.tick_params(bottom=False, left=False)
ax.legend(frameon=False)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(True, color='#EEEEEE')
ax.spines['bottom'].set_color('#DDDDDD')

fig.tight_layout()

plt.savefig('plots/Engines.jpg', dpi=400)
```
