```python
import json
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
```

## Here we do some quick EDA on the lichess data

We see that the distribution of elo ratings is approximately normal with mean around 1600.

```python
path = "..\data\Curated\lichess_2024.ndjson"
df = pd.read_json(path, lines=True)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['WhiteElo'], bins = 20, alpha=.7, edgecolor='black', label="White Elo Rating")
plt.hist(df['BlackElo'], bins = 20, alpha=.4, edgecolor='grey', label="Black Elo Rating")

plt.xlabel('Elo Rating')
plt.ylabel('Frequency')
plt.title('Distribution of play elo ratings')

plt.legend()

plt.grid(True)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.savefig('../plots/Distribution of elo ratings.png', dpi=400)
plt.show()
```
